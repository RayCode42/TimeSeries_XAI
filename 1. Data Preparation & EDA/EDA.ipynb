```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Define the list of tickers
tickers = ['BND_enrich.csv', 'GLD_enrich.csv', 'GOVT_enrich.csv', 'IVOO_enrich.csv', 
           'JNK_enrich.csv', 'SPY_enrich.csv', 'VIOO_enrich.csv', 'VOO_enrich.csv']

# Function to generate the plots
def plot_time_series_analysis(ticker):
    # Read the dataset
    df = pd.read_csv(ticker, parse_dates=['Date'], index_col='Date')
    ticker = ticker.split(".")[0].split("_")[0]

    # Create a new figure
    plt.figure(figsize=(10, 8))

    # Time series plot
    plt.subplot(2, 2, 1)
    plt.plot(df.index, df['Close'], label='Closing Price')
    plt.plot(df.index, df['Close'].rolling(window=30).mean(), label='30 Day MA', color='blue')
    plt.title(f'{ticker} Closing Price')
    plt.xlabel('')
    plt.ylabel('Closing Price')
    plt.legend()

    # ACF plot
    plt.subplot(2, 2, 2)
    plot_acf(df['Close'], lags=35, ax=plt.gca())
    plt.title('Sample ACF')

    # PACF plot
    plt.subplot(2, 2, 3)
    plot_pacf(df['Close'], lags=35, ax=plt.gca())
    plt.title('Sample PACF')

    # Density plot
    plt.subplot(2, 2, 4)
    sns.kdeplot(df['Close'], shade=True)
    plt.title('Density')
    plt.xlabel('Closing Price')

    # Adjust layout
    plt.tight_layout()

    # Save the plot
    plt.savefig(f'{ticker}_time_series_analysis.png')
    plt.show()

# Loop through all tickers and generate the plots
for ticker in tickers:
    plot_time_series_analysis(ticker)
```
